```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
from astropy.wcs import WCS

from astropy.coordinates import SkyCoord
import astropy.units as u

from datetime import timedelta
import pandas as pd
```

```python
def raAndDec(data):
    posList = ['x','y','z']
    earthMoonDist = 384000#km (avg distance)
    earthUnitVector = np.zeros((len(data),len(posList)))
    pos = np.zeros((len(data),len(posList)))
    for i in range(0,len(data)):
        for j in range(0,len(posList)):
            earthUnitVector[i][j] = data.iloc[i]['earth_unit_vector_'+posList[j]]#pointing from center of moon
            pos[i][j] = data.iloc[i]['position_'+posList[j]]
    earthVector = earthUnitVector*earthMoonDist
    raeVector = -earthVector+pos#should be earth centered position of rae
    ra,dec = cartToSphere(raeVector[:,0],raeVector[:,1],raeVector[:,2])
    return ra,dec
```

```python
def cartToSphere(x, y, z):
    # Calculate radius
    r = np.sqrt(x**2 + y**2 + z**2)
    
    # Calculate declination (latitude in spherical terms)
    dec = np.arcsin(z / r)
    
    # Calculate right ascension (longitude in spherical terms)
    ra = np.arctan2(y, x)
    
    # Convert to degrees
    dec = np.degrees(dec)
    ra = np.degrees(ra)
    
    # Ensure ra is in the range [0, 360]
    ra = ra % 360
    
    return ra, dec
```

```python
# Load the file to inspect its contents
filename = r"C:\Users\adamf\Documents\PhD\Diffraction\interpolatedRAE2MasterFile.csv"

rawData = pd.read_csv(filename)
data = rawData
data['time'] = pd.to_datetime(data['time'])
data.drop('Unnamed: 0',axis=1,inplace=True)
data.set_index('time',inplace=True)
```

```python
start_date = pd.to_datetime("1974-02-27 00:30")
end_date = pd.to_datetime("1974-02-27 22:00")

# Selecting rows within the date range
earthOccult = data[(data.index >= start_date) & (data.index <= end_date)].copy()
```

```python
earthOccult['ra'],earthOccult['dec'] = raAndDec(earthOccult)
```

```python
ra_points = earthOccult['right_ascension'].values*15
dec_points = earthOccult['declination'].values
```

```python
with fits.open('Maps/haslam408_dsds_Remazeilles2014.fits') as hdul:
    print(hdul.info())  # Display the structure of the FITS file
    
    # Access the data from the Binary Table HDU (hdul[1])
    data_table = hdul[1].data
    print("Column names:", data_table.columns.names)
    
    # Extract the first column of the table, which contains the map data
    data_array = data_table.field(0)
    print("Data array shape:", data_array.shape)
```

```python
mean = np.mean(data_array)
std = np.std(data_array)
threshold = mean + 3 * std  # Set threshold as 3 standard deviations from the mean
data_array = np.where(data_array > threshold, np.nan, data_array)
```

```python
nlat, nlon = data_array.shape
lons = np.linspace(0, 360, nlon)  # Galactic longitude from 0 to 360 degrees
lats = np.linspace(-90, 90, nlat)
```

```python
lons_grid, lats_grid = np.meshgrid(lons, lats)
galactic_coords = SkyCoord(l=lons_grid * u.deg, b=lats_grid * u.deg, frame='galactic')
equatorial_coords = galactic_coords.transform_to('icrs')
ra_grid = equatorial_coords.ra.deg
dec_grid = equatorial_coords.dec.deg
```

```python
plt.figure(figsize=(14, 7))  # Rectangular figure size
plt.imshow(data_array, extent=[np.min(ra_grid), np.max(ra_grid), np.min(dec_grid), np.max(dec_grid)],
           cmap='inferno', origin='lower', aspect='auto')  # Adjusted aspect ratio
plt.colorbar(label="Brightness Temperature (K)")
plt.title("Haslam 408 MHz Sky Map with RA/Dec Grid")
plt.xlabel("Right Ascension (degrees)")
plt.ylabel("Declination (degrees)")
plt.gca().set_aspect(1.5)
ra_lines = np.linspace(0, 360, 13)  # RA lines every 30 degrees
dec_lines = np.linspace(-90, 90, 7)  # Dec lines every 30 degrees
for ra in ra_lines:
    plt.plot([ra, ra], [-90, 90], color='white', linestyle='--', linewidth=0.5)
for dec in dec_lines:
    plt.plot([0, 360], [dec, dec], color='white', linestyle='--', linewidth=0.5)
plt.scatter(ra_points, dec_points, color='red', marker='x', label='RA/Dec Points')
```
